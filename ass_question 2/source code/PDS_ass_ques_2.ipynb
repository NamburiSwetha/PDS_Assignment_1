```python
import pandas as pd
data=pd.read_csv('StudentsPerformance.csv')
data.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['reading score'],data['math score'])
plt.show()
```

```python
info=data[["test preparation course","math score","reading score","writing score"]]
average_scores=info.groupby("test preparation course").mean()
average_scores
average_scores = average_scores.reset_index()
sns.barplot(x='test preparation course', y='value', hue='variable', 
            data=pd.melt(average_scores, id_vars='test preparation course'), 
            palette='rocket')
plt.title('Average Scores by test preparation course')
plt.xlabel('test preparation course')
plt.ylabel('Average Score')
plt.ylim(0, 100)
plt.legend(title='Subject')
plt.tight_layout()
```

```python

import seaborn as sns
sns.boxplot(x='race/ethnicity', y='reading score',  data=data)
```

```python
parent_edu_scores = data.groupby('parental level of education').mean()[['math score', 'reading score', 'writing score']]
sns.heatmap(parent_edu_scores, cmap='coolwarm', annot=True, fmt='.1f')
plt.title('Correlation between Scores and Parental Education Levels')
plt.xlabel('Subject')
plt.ylabel('Parental Level of Education')
plt.show()

```

```python

sns.histplot(x='parental level of education', data=data, 
             hue='gender', multiple='dodge', shrink=0.8, 
             palette='rocket')
plt.title('Distribution of Parental Education Levels')
plt.xlabel('Parental Level of Education')
plt.ylabel('Count')
plt.tight_layout()

```
